```python
# Import dependencies
import pandas as pd
import os
```

```python
file_path = ['trip_data/JC-201801-citibike-tripdata.csv',
        'trip_data/JC-201802-citibike-tripdata.csv',
        'trip_data/JC-201803-citibike-tripdata.csv']
        
```

```python
citibike_df = pd.DataFrame()
```

```python
# Loop through the list of files and concatonate all data to the dataframe
for file in file_path:
    data = pd.read_csv(file)
    citibike_df = pd.concat([citibike_df, data], axis=0, ignore_index=True)
```

```python
citibike_df
```

```python
# Check for missing values in each column
missing_values = citibike_df.isnull().sum()
print(missing_values)
```

```python
# Store the dataframe as CSV file
citibike_df.to_csv('trip_data/2018jc_data.csv', index=False)
```
